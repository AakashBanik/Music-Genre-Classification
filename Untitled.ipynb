```python
import librosa
import matplotlib.pyplot as plt
import librosa.display
```

```python
x, sr = librosa.load('./1.mp3') #default sr=22khz, change sr for a variable sample rate, sr=None for no resampling
type(x)
```

```python
%matplotlib inline
plt.figure(figsize=(14,5))
librosa.display.waveplot(x, sr=sr) #show the plot of the mp3 file exported
```

```python
X = librosa.stft(x)
Xdb = librosa.amplitude_to_db(abs(X))
plt.figure(figsize=(14,5))
librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='hz')
plt.colorbar()
```

```python
librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='log') #as everything is in the lower spectrum, to view this properly, convert to log
plt.colorbar()
```

```python
librosa.output.write_wav('example.wav', x, sr) #write the numpy array to wav file
```
